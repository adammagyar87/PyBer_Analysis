### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df = pyber_data_df.set_index('type')
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_city_type = pyber_data_df.groupby('type').count()[['ride_id']]
rides_city_type = rides_city_type.reset_index()
rides_city_type
```

```python
# 2. Get the total drivers for each city type

# In Step 2, use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the sum() method to the "driver_count" column.

drivers_city_type = city_data_df.groupby(["type"]).sum()[["driver_count"]]
drivers_city_type

# This gives wrong values when using "pyber_data_df"
```

```python
#  3. Get the total amount of fares for each city type
fares_city_type = pyber_data_df.groupby('type').sum()[['fare']]
fares_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
# In Step 4, calculate the average fare per ride by city type by dividing the sum of all the fares by the total rides.

average_fare_per_ride = (fares_city_type / rides_city_type)
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
# In Step 5, calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers.

average_fare_per_driver = fares_city_type / drivers_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
        [{"Total Rides": rides_city_type,
          "Total Drivers": drivers_city_type,
          "Total Fares": fares_city_type,
          "Average Fare Per Ride": average_fare_per_ride,
          "Average Fare Per Driver":average_fare_per_driver}])
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
city_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

#create a new DataFrame with multiple indices using the groupby() function on the "type" and "date" columns of the
#pyber_data_df DataFrame, then apply the sum() method on the "fare" column to show the total fare amount for each date.

group_df = pyber_data_df.groupby(['type', 'date']).sum('fare')
group_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
group_df = group_df.reset_index()
group_df
```

```python
# Create a pivot table with the 'stoptime' as the index and the columns ='gender' with the trip counts in each row.
#gender_stoptime_pivot = gender_stoptime.pivot(index = 'stoptime', columns = 'gender', values= 'tripduration')
#gender_stoptime_pivot
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
group_df_pivot = group_df.pivot(index = 'date', columns = 'type', values = 'fare')
group_df_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_group_df_pivot = group_df_pivot.loc['2019-01-01':'2019-04-29', :]
new_group_df_pivot
```

```python
new_group_df_pivot.dtypes
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_group_df_pivot.index = pd.to_datetime(new_group_df_pivot.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_group_df_pivot.dtypes
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_rides_df = new_group_df_pivot.resample('W').count
weekly_rides_df.head()
```
